```python
from sunpycube.cube.datacube import Cube, CubeSequence
from astropy.io import fits
from sunpycube.wcs_util import WCS
import sunpycube.cube.cube_utils as cu
import os

file_location = "/home/ankit/Desktop/DATA/STEVE/"
files = os.listdir(file_location)

for i, f in enumerate(files):
	files[i] = file_location + f

hdulist = fits.open(files[0])
header = hdulist[1].header
wt = WCS(header)
data = hdulist[1].data
cube1 = Cube(data, wt)

hdulist = fits.open(files[1])
header = hdulist[1].header
wt = WCS(header)
data = hdulist[1].data
cube2 = Cube(data, wt)

hdulist = fits.open(files[2])
header = hdulist[1].header
wt = WCS(header)
data = hdulist[1].data
cube3 = Cube(data, wt)

hdulist = fits.open(files[3])
header = hdulist[1].header
wt = WCS(header)
data = hdulist[1].data
cube4 = Cube(data, wt)

seq = CubeSequence([cube1, cube2, cube3, cube4], common_axis=0)
```
