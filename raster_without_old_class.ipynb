```python
from sunpycube.cube.datacube import Cube
from sunpycube.wcs_util import WCS
import matplotlib.pyplot as plt
from astropy.io import fits
import astropy.units as u
raster_file = "/home/ankit/Desktop/DATA/STEVE/iris_l2_20141211_191222_3803105278_raster_t000_r00000.fits"
hdulist = fits.open(raster_file)
w = WCS(header=hdulist[1].header)
dc = Cube(hdulist[1].data, w)
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], projection=dc.axes_wcs, slices=(0, 'y',  'x'))
ax.coords[2].set_ticks(exclude_overlapping=True)
ax.imshow(dc.data[0, :, :], cmap=plt.cm.gist_heat, origin='lower')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8], projection=dc.axes_wcs, slices=(0, 'y',  'x'))
print(type(ax))
ax.coords[2].set_ticks(exclude_overlapping=True)
ax.coords[2].set_format_unit(u.nm)

data = dc.data[0]
im = plt.imshow(data, cmap=plt.cm.gist_heat, origin='lower', animated=True)


def updatefig(i):
    global data
    if i==6:
        ax.reset_wcs(wcs=w, slices=(0, 'x',  'y'))
    data = dc.data[i]
    im.set_array(data)
    return im,

anim = animation.FuncAnimation(fig, updatefig,frames=dc.data.shape[0], interval=200, blit=True)

HTML(anim.to_html5_video())
```
